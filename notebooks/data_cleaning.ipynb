# Clean all 10 .csv file

## 1. Clean 01_fund_master.csv

```python
# Remove duplicates
fund_master = fund_master.drop_duplicates()

# Remove spaces
text_cols = ['fund_house','scheme_name','category','sub_category',
             'plan','benchmark','fund_manager',
             'risk_category','sebi_category_code']

for col in text_cols:
    fund_master[col] = fund_master[col].astype(str).str.strip()

# Convert date
fund_master['launch_date'] = pd.to_datetime(
    fund_master['launch_date'],
    errors='coerce'
)

# Numeric columns
numeric_cols = [
    'expense_ratio_pct',
    'exit_load_pct',
    'min_sip_amount',
    'min_lumpsum_amount'
]

for col in numeric_cols:
    fund_master[col] = pd.to_numeric(
        fund_master[col],
        errors='coerce'
    )

# Remove duplicate AMFI codes
fund_master = fund_master.drop_duplicates(subset='amfi_code')

# Validate expense ratio
fund_master = fund_master[
    fund_master['expense_ratio_pct'].between(0.1,2.5)
]

fund_master.to_csv(
    "../data/processed/clean_fund_master.csv",
    index=False
)
```

## 2. Clean 02_nav_history.csv

```python
nav_history['date'] = pd.to_datetime(
    nav_history['date'],
    errors='coerce'
)

nav_history = nav_history.sort_values(
    ['amfi_code','date']
)

nav_history = nav_history.drop_duplicates()

nav_history['nav'] = (
    nav_history.groupby('amfi_code')['nav']
    .ffill()
)

nav_history = nav_history[
    nav_history['nav'] > 0
]

nav_history.to_csv(
    "../data/processed/clean_nav_history.csv",
    index=False
)
```

## 3. Clean 03_aum_by_fund_house.csv

```python
aum = aum.drop_duplicates()

aum['date'] = pd.to_datetime(
    aum['date'],
    errors='coerce'
)

aum['fund_house'] = (
    aum['fund_house']
    .str.strip()
)

numeric_cols = [
    'aum_lakh_crore',
    'aum_crore',
    'num_schemes'
]

for col in numeric_cols:
    aum[col] = pd.to_numeric(
        aum[col],
        errors='coerce'
    )

    aum = aum[aum[col] >= 0]

aum.to_csv(
    "../data/processed/clean_aum_by_fund_house.csv",
    index=False
)
```

## 4. Clean 04_monthly_sip_inflows.csv

```python
sip = sip.drop_duplicates()

sip['month'] = pd.to_datetime(
    sip['month'],
    errors='coerce'
)

cols = [
    'sip_inflow_crore',
    'active_sip_accounts_crore',
    'new_sip_accounts_lakh',
    'sip_aum_lakh_crore',
    'yoy_growth_pct'
]

for col in cols:
    sip[col] = pd.to_numeric(
        sip[col],
        errors='coerce'
    )

sip.to_csv(
    "../data/processed/clean_monthly_sip_inflows.csv",
    index=False
)
```

## 5. Clean 05_category_inflows.csv

```python
category = category.drop_duplicates()

category['month'] = pd.to_datetime(
    category['month'],
    errors='coerce'
)

category['category'] = (
    category['category']
    .str.strip()
)

category['net_inflow_crore'] = pd.to_numeric(
    category['net_inflow_crore'],
    errors='coerce'
)

category.to_csv(
    "../data/processed/clean_category_inflows.csv",
    index=False
)
```

## 6. Clean 06_industry_folio_count.csv

```python
folio = folio.drop_duplicates()

folio['month'] = pd.to_datetime(
    folio['month'],
    errors='coerce'
)

cols = [
    'total_folios_crore',
    'equity_folios_crore',
    'debt_folios_crore',
    'hybrid_folios_crore',
    'others_folios_crore'
]

for col in cols:
    folio[col] = pd.to_numeric(
        folio[col],
        errors='coerce'
    )

folio.to_csv(
    "../data/processed/clean_industry_folio_count.csv",
    index=False
)
```

## 7. Clean 07_scheme_performance.csv

```python
performance = performance.drop_duplicates()

numeric_cols = [
    'return_1yr_pct',
    'return_3yr_pct',
    'return_5yr_pct',
    'benchmark_3yr_pct',
    'alpha',
    'beta',
    'sharpe_ratio',
    'sortino_ratio',
    'std_dev_ann_pct',
    'max_drawdown_pct',
    'aum_crore',
    'expense_ratio_pct'
]

for col in numeric_cols:
    performance[col] = pd.to_numeric(
        performance[col],
        errors='coerce'
    )

# Flag negative Sharpe
negative_sharpe = performance[
    performance['sharpe_ratio'] < 0
]

print(negative_sharpe)

performance = performance[
    performance['expense_ratio_pct'].between(0.1,2.5)
]

performance.to_csv(
    "../data/processed/clean_scheme_performance.csv",
    index=False
)
```

## 8. Clean 08_investor_transactions.csv

```python
transactions = transactions.drop_duplicates()

transactions['transaction_date'] = pd.to_datetime(
    transactions['transaction_date'],
    errors='coerce'
)

transactions['transaction_type'] = (
    transactions['transaction_type']
    .str.strip()
    .str.title()
)

transactions = transactions[
    transactions['transaction_type'].isin(
        ['Sip','Lumpsum','Redemption']
    )
]

transactions = transactions[
    transactions['amount_inr'] > 0
]

transactions = transactions[
    transactions['kyc_status'].isin(
        ['Verified','Pending']
    )
]

transactions.to_csv(
    "../data/processed/clean_investor_transactions.csv",
    index=False
)
```

## 9. Clean 09_portfolio_holdings.csv

```python
holdings = holdings.drop_duplicates()

holdings['portfolio_date'] = pd.to_datetime(
    holdings['portfolio_date'],
    errors='coerce'
)

cols = [
    'weight_pct',
    'market_value_cr',
    'current_price_inr'
]

for col in cols:
    holdings[col] = pd.to_numeric(
        holdings[col],
        errors='coerce'
    )

holdings = holdings[
    holdings['weight_pct'].between(0,100)
]

holdings.to_csv(
    "../data/processed/clean_portfolio_holdings.csv",
    index=False
)
```

## 10. Clean 10_benchmark_indices.csv

```python
benchmark = benchmark.drop_duplicates()

benchmark['date'] = pd.to_datetime(
    benchmark['date'],
    errors='coerce'
)

benchmark = benchmark.sort_values('date')

numeric_cols = benchmark.columns.drop('date')

for col in numeric_cols:
    benchmark[col] = pd.to_numeric(
        benchmark[col],
        errors='coerce'
    )

benchmark = benchmark.ffill()

benchmark.to_csv(
    "../data/processed/clean_benchmark_indices.csv",
    index=False
)
```

# Create SQLite Database

```python
engine = create_engine("sqlite:///../data/bluestock_mf.db")

print("Database Created Successfully!")
```

```python
fund_master.to_sql(
    "dim_fund",
    engine,
    if_exists="replace",
    index=False
)

nav_history.to_sql(
    "fact_nav",
    engine,
    if_exists="replace",
    index=False
)

aum.to_sql(
    "fact_aum",
    engine,
    if_exists="replace",
    index=False
)

sip.to_sql(
    "fact_sip_industry",
    engine,
    if_exists="replace",
    index=False
)

category.to_sql(
    "fact_category_inflows",
    engine,
    if_exists="replace",
    index=False
)

folio.to_sql(
    "fact_folio_count",
    engine,
    if_exists="replace",
    index=False
)

performance.to_sql(
    "fact_performance",
    engine,
    if_exists="replace",
    index=False
)

transactions.to_sql(
    "fact_transactions",
    engine,
    if_exists="replace",
    index=False
)

holdings.to_sql(
    "fact_portfolio",
    engine,
    if_exists="replace",
    index=False
)

benchmark.to_sql(
    "dim_benchmark",
    engine,
    if_exists="replace",
    index=False
)

print("All Tables Loaded Successfully!")
```
