# Clean NAV History

```python
nav_history['date'] = pd.to_datetime(nav_history['date'])

nav_history = nav_history.sort_values(
    ['amfi_code','date']
)

nav_history = nav_history.drop_duplicates()

nav_history['nav'] = (
    nav_history.groupby('amfi_code')['nav']
    .ffill()
)

nav_history = nav_history[
    nav_history['nav'] > 0
]
```

```python
nav_history.info()
```

# Clean Transactions

```python
transactions['transaction_date'] = pd.to_datetime(
    transactions['transaction_date']
)

transactions = transactions[
    transactions['amount_inr'] > 0
]

transactions['transaction_type'] = (
    transactions['transaction_type']
    .str.strip()
)
```

```python
transactions.info()
```

# Clean Performance

```python
performance['return_1yr_pct'] = pd.to_numeric(
    performance['return_1yr_pct'],
    errors='coerce'
)

performance['sharpe_ratio'] = pd.to_numeric(
    performance['sharpe_ratio'],
    errors='coerce'
)
```

```python
performance.info()
```

# Save Cleaned Files

```python
nav_history.to_csv(
    "../data/processed/clean_nav.csv",
    index=False
)

transactions.to_csv(
    "../data/processed/clean_transactions.csv",
    index=False
)

performance.to_csv(
    "../data/processed/clean_performance.csv",
    index=False
)
```

# Create Database

```python
engine = create_engine(
    "sqlite:///bluestock_mf.db"
)
```

# Load Tables

```python
fund_master.to_sql(
    "dim_fund",
    engine,
    if_exists="replace",
    index=False
)

nav_history.to_sql(
    "fact_nav",
    engine,
    if_exists="replace",
    index=False
)

transactions.to_sql(
    "fact_transactions",
    engine,
    if_exists="replace",
    index=False
)

performance.to_sql(
    "fact_performance",
    engine,
    if_exists="replace",
    index=False
)
```
